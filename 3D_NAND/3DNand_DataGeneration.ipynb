```python
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d

# Function to interpolate and return DataFrame
def interpolate_data(x, y, x_label, y_label, kind='cubic', points=100):
    interpolator = interp1d(x, y, kind=kind)
    x_interpolated = np.linspace(x.min(), x.max(), points)
    y_interpolated = interpolator(x_interpolated)
    return pd.DataFrame({x_label: x_interpolated, y_label: y_interpolated})

# 1. Gate Length Effect on Threshold Voltage Shift
Lg = np.array([70, 90, 130, 150, 210, 300])  # Gate lengths
Vt = np.array([1.746, 2.081, 2.334, 2.429, 2.583, 2.666])  # Threshold voltages
gate_length_data = interpolate_data(Lg, Vt, "Gate Length (nm)", "Threshold Voltage (V)")

# 2. Positive Gate Voltage Effect on Threshold Voltage Shift (Program Voltage)
Vp = np.array([8, 9, 10, 11, 12, 13])  # Positive gate voltages
dVt_gate = np.array([0.611, 0.727, 0.789, 0.798, 0.799, 0.799])  # Threshold voltage shifts
gate_voltage_data = interpolate_data(Vp, dVt_gate, "Gate Voltage (V)", "Threshold Voltage Shift (Vt)")

# 3. Threshold Voltage Dependence on Negative Gate Voltage (Erase Voltage)
Ve = np.array([-8, -9, -10, -11, -12, -13])  # Negative gate voltages
dVt_negative = np.array([-0.361, -1.111, -1.199, -1.199, -1.199, -1.199])  # Threshold voltage shifts
negative_gate_data = interpolate_data(Ve, dVt_negative, "Negative Gate Voltage (V)", "Threshold Voltage Shift (Vt)")

# 4. Temperature Effect on Threshold Voltage Shift (Celsius)
T = np.array([-23, 23, 27, 75, 85, 100])  # Temperatures in Celsius
dVt_temp = np.array([1.489, 1.481, 1.48, 1.472, 1.47, 1.46])  # Threshold voltage shifts
temperature_data = interpolate_data(T, dVt_temp, "Temperature (C)", "Threshold Voltage Shift (Vt)")

# 6. Dependence of Memory Window on Si3N4 Layer
nitride_thickness_window = np.array([2, 2.5, 3.5, 4.5, 5.5, 6, 7, 8])  # Nitride thickness in nm
threshold_voltage_shift_window = np.array([0.1, 0.15, 0.4, 0.62, 0.82, 0.84, 1.12, 1.52])  # Voltage shifts
memory_window_data = interpolate_data(nitride_thickness_window, threshold_voltage_shift_window,
                                      "Nitride Thickness (nm)", "Threshold Voltage Shift (Vth)")


# Exporting to Excel
excel_path = "NandData1.xlsx"
with pd.ExcelWriter(excel_path, engine='openpyxl') as writer:
    # Write each analysis to a separate sheet
    gate_length_data.to_excel(writer, sheet_name="Gate Length vs Vt", index=False)
    gate_voltage_data.to_excel(writer, sheet_name="Gate Voltage vs Vt", index=False)
    negative_gate_data.to_excel(writer, sheet_name="Negative Gate Voltage vs Vt ", index=False)
    temperature_data.to_excel(writer, sheet_name="Temperature vs Vt", index=False)
    memory_window_data.to_excel(writer, sheet_name="Memory Window vs Si3N4", index=False)

print(f"Data has been exported to {excel_path}")

```

```python
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d

# Function to interpolate and return DataFrame
def interpolate_data(x, y, x_label, y_label, kind='cubic', points=100):
    interpolator = interp1d(x, y, kind=kind, bounds_error=False, fill_value="extrapolate")
    x_interpolated = np.linspace(x.min(), x.max(), points)  # Ensure interpolation range is valid
    y_interpolated = interpolator(x_interpolated)
    return pd.DataFrame({x_label: x_interpolated, y_label: y_interpolated})

# 5. Effect of Different Tunnel Oxide on Threshold Voltage Shift
Vp = np.array([7, 8, 9, 10, 11, 12])  # Gate voltages
dVth_data = {
    "0.5nm": np.array([-0.351, -0.187, -0.052, 0.075, 0.076, 0.076]),
    "1nm": np.array([np.nan, 0.072, 0.209, 0.263, 0.263, 0.263]),  # 'Failed' treated as NaN
    "1.5nm": np.array([0.096, 0.271, 0.397, 0.428, 0.428, 0.428]),
    "2nm": np.array([0.363, 0.471, 0.567, 0.6, 0.6, 0.6]),
    "2.5nm": np.array([0.594, 0.614, 0.612, 0.59, 0.567, 0.555]),
    "3nm": np.array([0.714, 0.716, 0.715, 0.708, 0.694, 0.682]),
    "3.5nm": np.array([0.283, 0.45, 0.715, 0.713, 0.703, 0.691]),
}

# Interpolating the data for each tunnel oxide thickness
interpolated_tunnel_data1 = {}
Vp_interpolated = np.linspace(Vp.min(), Vp.max(), 100)  # Interpolated Vp values
for oxide_thickness, dVth in dVth_data.items():
    # Exclude NaN values for interpolation
    valid_indices = ~np.isnan(dVth)
    Vp_valid = Vp[valid_indices]
    dVth_valid = dVth[valid_indices]

    # Interpolation for each tunnel oxide thickness
    interpolator = interp1d(Vp_valid, dVth_valid, kind='cubic', bounds_error=False, fill_value="extrapolate")
    dVth_interpolated = interpolator(Vp_interpolated)

    # Store interpolated data for each oxide thickness
    interpolated_tunnel_data1[oxide_thickness] = dVth_interpolated

# 7. Nitride Thickness vs Threshold Voltage Shift at Different Tunnel Oxide Thicknesses
nitride_thickness = np.array([2, 2.45, 3.5, 4.5, 5.7, 6, 7, 8])  # Nitride thickness in nm
dVth_data2 = {
    "1.8nm": np.array([0.91, 0.93, 0.98, 1.15, 1.10, 1.12, 1.17, 1.22]),
    "2nm": np.array([0.93, 0.97, 1.01, 1.17, 1.12, 1.14, 1.20, 1.25]),
    "2.5nm": np.array([0.95, 1.01, 1.06, 1.12, 1.16, 1.19, 1.25, 1.30]),
    "3nm": np.array([1.04, 1.06, 1.12, 1.17, 1.22, 1.25, 1.30, 1.35]),
    "4nm": np.array([1.14, 1.17, 1.22, 1.27, 1.33, 1.36, 1.40, 1.45]),
}

# Interpolating the data for each tunnel oxide thickness
interpolated_tunnel_data2 = {}
nitride_thickness_interpolated = np.linspace(nitride_thickness.min(), nitride_thickness.max(), 100)  # Interpolated Nitride thickness values
for oxide_thickness, dVth in dVth_data2.items():
    # Interpolation for each tunnel oxide thickness
    interpolator = interp1d(nitride_thickness, dVth, kind='cubic', bounds_error=False, fill_value="extrapolate")
    dVth_interpolated = interpolator(nitride_thickness_interpolated)

    # Store interpolated data for each oxide thickness
    interpolated_tunnel_data2[oxide_thickness] = dVth_interpolated

# Exporting to Excel
excel_path = "NandData2.xlsx"
with pd.ExcelWriter(excel_path, engine='openpyxl') as writer:
    # 5. Effect of Different Tunnel Oxide on Threshold Voltage Shift (in the same table)
    # Create the base DataFrame with the first column as Vp_interpolated
    effect_tunnel_oxides = pd.DataFrame({"Vp (V)": Vp_interpolated})

    # Add the interpolated data for each oxide thickness as separate columns
    for oxide_thickness, dVth in interpolated_tunnel_data1.items():
        effect_tunnel_oxides[f"Vth ({oxide_thickness})"] = dVth

    # Write to the Excel file
    effect_tunnel_oxides.to_excel(writer, sheet_name="Effect of Tunnel Oxide on Vth", index=False)

    # 7. Nitride Thickness vs Threshold Voltage Shift at Different Tunnel Oxide Thicknesses (in the same table)
    # Create the base DataFrame with the first column as Nitride_thickness_interpolated
    nitride_tunnel_oxides = pd.DataFrame({"Nitride Thickness (nm)": nitride_thickness_interpolated})

    # Add the interpolated data for each oxide thickness as separate columns
    for oxide_thickness, dVth in interpolated_tunnel_data2.items():
        nitride_tunnel_oxides[f"Vth ({oxide_thickness})"] = dVth

    # Write to the Excel file
    nitride_tunnel_oxides.to_excel(writer, sheet_name="Nitride Thickness vs Vth", index=False)

print(f"Data has been exported to {excel_path}")
```
